### Character Arcs of Karna v/s Arjun

```python
#importing necessary libraries
import io
import pandas as pd
import numpy as np
from pprint import pprint
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#setting graph size
fig_size=[10,5]
plt.rcParams["figure.figsize"] = fig_size
```

```python
#reading the megaepic
with io.open ("../Data/mhbhrt/mahafull.txt", encoding ='utf8') as f_in:
    mhb_text = f_in.read()
```

```python
mhb_books = mhb_text.split('BOOK')
# BOOK Split as Mahabharata has 18 books.
```

```python
# Getting the actual 18 books from the epic.
mhb_book_list = mhb_books[1:]
```

```python
# Extracting Book Title
book_titles=[]
for i in range(len(mhb_book_list)):
    book_title = mhb_book_list[i].split("\n")[2].strip()
    book_titles.append(book_title)
book_titles
```

```python
characters = ["Arjuna", "Partha", "Dhananjaya", "Gudakesa", "Phalguni", "Vijaya","Savyasachin","Vibhatsu","Kiritin","Jishnu"]
```

```python
characters2 = ["Vasusena", "Suta's son","Karna","Radheya","Vrisha"]
```

```python
# Counting the occurences of names of Arjun in the epic
counts = ["Book" + '\t' + '\t'.join(characters)]
for i in range(len(mhb_book_list)):
    line = book_titles[i]
    for ch in characters:
        line = line + '\t' + str(mhb_book_list[i].count(ch))
    counts.append(line) 
```

```python
# type(counts)
counts
```

```python
with open("../Data/mhb_arjun.txt",'w') as f_out:
    for line in counts:
        f_out.write(line+'\n')
```

```python
arj_df = pd.read_csv("../Data/mhb_arjun.txt", sep = '\t')
# arj_df=arj_df.set_index(['Book'])
arj_df['Arjun_names'] = arj_df.sum(1)
arj_df
```

```python
arj_df.plot(legend=True)
```

```python
#Counting the occurences of names of Karna in the epic
counts2 = ["Book" + '\t' + '\t'.join(characters2)]
for i in range(len(mhb_book_list)):
    line = book_titles[i]
    for ch in characters2:
        line = line + '\t' + str(mhb_book_list[i].count(ch))
    counts2.append(line) 
```

```python
with open("../Data/mhb_karna.txt",'w') as f_out:
    for line in counts2:
        f_out.write(line+'\n')
```

```python
krn_df = pd.read_csv("../Data/mhb_karna.txt", sep = '\t')
# krn_df=krn_df.set_index(['Book'])
krn_df['Karna_names'] = krn_df.sum(1)
# type(krn_df)
krn_df
```
